```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline
```

```python
x = pd.read_csv('F:\GITHUB\Python\Diamond_Cuts.csv')
```

```python
x.head(3)
```

```python
g1 = x.loc[x['cut']=='Ideal','depth']
g2 = x.loc[x['cut']=='Premium','depth']
g3 = x.loc[x['cut']=='Good','depth']
g4 = x.loc[x['cut']=='Fair','depth']
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(10,8),dpi=80)
plt.hist(g1,color='g',label='IDEAL',alpha= 0.5,bins = 100)
plt.hist(g2,color='r',label='PREMIUM',alpha= 0.5,bins = 100)
plt.hist(g3,color='b',label='GOOD',alpha= 0.5,bins = 100)
plt.legend(fontsize=10)
plt.title("Diamond Depth's Histogram")
plt.ylabel('Frequency')
plt.xlim(55,70)
plt.show()
```

#### Subplots 

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(10,5),sharey=True)
ax1.hist(g1,color='g',label='IDEAL',alpha= 0.5,bins = 100)
ax2.hist(g2,color='r',label='PREMIUM',alpha= 0.5,bins = 100)
ax3.hist(g3,color='b',label='GOOD',alpha= 0.5,bins = 100)
for i in ax1,ax2,ax3:
    i.legend()
plt.show()
```

#### Seaborn Density Curve 

```python
import seaborn as sns

fig,axes = plt.subplots(1,4,figsize=(10,5),sharey=True)
sns.distplot(g1,color='green',label='IDEAL',ax = axes[0])
sns.distplot(g2,color='gold',label='PREMIUM',ax = axes[1])
sns.distplot(g3,color='skyblue',label='GOOD',ax = axes[2])
sns.distplot(g4,color='deeppink',label='FAIR',ax = axes[3])

for i in axes:
    i.legend()
plt.xlim(50,80)
plt.show()
```
